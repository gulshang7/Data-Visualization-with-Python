## Line Plot

```python
df= pd.DataFrame(np.random.randn(100,3), columns = ['A', 'B', 'C'])
df.head()
```

```python
df = df.cumsum()
df.head()
```

```python
df.iplot(title = 'Line Plot')
```

```python
df.plot()
```

## Scatter Plot

```python
df.iplot(x = 'A', y = 'B', mode = 'markers', size = 20)
```

```python
df.iplot(x = 'A', y = 'B', mode = 'markers', size = 20, title = 'Scatter Plot', xTitle = 'X-Axis', yTitle = 'Y-Axis')
```

## Bar Plot

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.iplot(kind = 'bar', x = 'sex', y= 'survived', title = 'Titanic')
```

```python
titanic.iplot(kind = 'bar', x = 'embark_town', y= 'survived', title = 'Titanic', xTitle = 'Town', yTitle = 'Survived')
```

```python
df.iplot(kind = 'bar')
```

```python
df.iplot(kind = 'bar', barmode = 'stack', bargap = 0.5, title = 'Bar Plot')
```

```python
df.iplot(kind = 'barh', barmode = 'stack', bargap = 0.5)
```

## Box Plot and Area Plot

```python
df.iplot(kind = 'box', title = 'Box Plot')
```

```python
df.iplot(kind = 'area', fill = True, title = 'Area Plot')
```

## 3D Plot

```python
df.iplot(kind = 'surface', title = '3D Plot')
```

```python
df.iplot(kind = 'surface', colorscale = 'rdylbu')
```

```python
cf.datagen.sinwave(10,0.25).iplot(kind = 'surface', title = '3D Plot')
```

## Spread Plot and Hist Plot

```python
df.iplot(kind = 'spread', title = 'Spread Plot')
```

```python
df.iplot(kind = 'hist', bins = 20, title = 'Histogram')
```

```python
df.iplot(kind = 'hist', bins = 20, barmode = 'overlay', bargap = 0.5, title = 'Histogram')
```

## Bubble Plot and Heatmap

```python
bb = cf.datagen.bubble3d(5,4, mode = 'stocks')
bb.head()
```

```python
bb.iplot(kind = 'bubble3d', x = 'x', y = 'y', z = 'z', size = 'size', title = 'Bubble 3D Plot')
```

```python
df.corr()
```

```python
df.corr().iplot(kind = 'heatmap')
```

```python
df.corr().iplot(kind = 'heatmap', colorscale = 'spectral')
```

```python
cf.datagen.heatmap(20,20).iplot(kind = 'heatmap', title = 'Heatmap')
```

```python
cf.datagen.heatmap(20,20).iplot(kind = 'heatmap', colorscale = 'spectral', title = 'Heatmap')
```

```python
cf.datagen.heatmap(20,20).iplot(kind = 'heatmap', colorscale = 'rdylbu')
```
