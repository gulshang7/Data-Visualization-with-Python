### Lambda Function

```python
x = lambda a: a*a*a
```

```python
x(3)
```

```python
mean_val = df['Mass'].mean()
```

```python
mean_val
```

```python
meta = df['Mass'].apply(lambda x: x if x==x else mean_val)
```

```python
meta.isnull().sum()
```

```python
df['Mass'] = meta
```

```python
df.isnull().sum()
```
