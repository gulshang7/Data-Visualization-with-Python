### Box Plot

```python
df.plot(kind = 'box', figsize=(10,5))
# -> Greean line is Mean, its divided 50-50% data
# -> at bottom line its Minimum value (0%) then 25%, then 50%, then 75% and then 100% i.e Maximum value
# -> little cicles are Outliers which is automatically detected by system 
```

```python
df.describe()
```

```python
df.plot(kind = 'box', figsize = (10,5), vert = False)
```

### Area and Scatter Plot

```python
df.plot(kind = 'area') #--> But it is stacked
```

```python
df.plot(kind = 'area', stacked= False)
```

```python
df.plot(kind = 'area', stacked = False, alpha = 0.9)
```

```python
df.plot.scatter(x = 'sepal_width', y= 'petal_width')
```

```python
df.plot.scatter(x = 'sepal_width', y= 'petal_width', c = 'sepal_length')
#--> If sepal_length is high points are DARKER and if low then LIGHTER
```

```python
ax = df.plot.scatter(x = 'sepal_width', y= 'petal_width')
df.plot.scatter(x = 'sepal_length', y= 'petal_length', ax = ax, color = 'r')
```

```python
ax = df.plot.scatter(x = 'sepal_width', y= 'petal_width', label = 'width')
df.plot.scatter(x = 'sepal_length', y= 'petal_length', ax = ax, color = 'r', label = 'length')
```

### Hex and Pie Plot

### Scatter matrix and subplots

```python
from pandas.plotting import scatter_matrix
```

```python
scatter_matrix(df, figsize = (8,8), diagonal = 'kde', color= 'r')
plt.show()
```

```python
df.plot(subplots = True)
plt.show()
```

```python
df.plot(subplots = True, sharex = False)
plt.tight_layout()
plt.show()
```

```python
df.plot(subplots = True, sharex = False, layout = (2,2), figsize = (6,6))
plt.tight_layout()
plt.show()
plt.savefig("Scatter matrix and subplots.jpg")
```
